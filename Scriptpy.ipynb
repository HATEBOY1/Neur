```python
!pip install scikit-learn

!wget https://github.com/HATEBOY1/Neur/blob/main/dataset.mat?raw=true
!wget https://github.com/HATEBOY1/Neur/blob/main/P1_0,275.png?raw=true

!mv dataset.mat\?raw\=true data.mat
!mv P1_0,275.png\?raw\=true p1.png

!pip install -U -q scipy

from scipy import io
import numpy as np
import scipy.io as sio 
import sklearn
from sklearn import svm
from scipy import io
from sklearn.model_selection import GridSearchCV, LeaveOneOut
import pylab as plt
from sklearn.preprocessing import StandardScaler
from matplotlib.colors import Normalize
from sklearn.metrics import mean_squared_error as mse
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
from PIL import Image


def make_colormap(seq):
    """Return a LinearSegmentedColormap
    seq: a sequence of floats and RGB-tuples. The floats should be increasing
    and in the interval (0,1).
    """
    seq = [(None,) * 3, 0.0] + list(seq) + [1.0, (None,) * 3]
    cdict = {'red': [], 'green': [], 'blue': []}
    for i, item in enumerate(seq):
        if isinstance(item, float):
            r1, g1, b1 = seq[i - 1]
            r2, g2, b2 = seq[i + 1]
            cdict['red'].append([item, r1, r2])
            cdict['green'].append([item, g1, g2])
            cdict['blue'].append([item, b1, b2])
    return mcolors.LinearSegmentedColormap('CustomMap', cdict)


def scaled_score(score,n_class):
    return (n_class/2*(score+1)-1)/(n_class-1)

n_class=10

mat_contents = io.loadmat('data.mat')

"""mat_contents = io.loadmat('HATEBOY1/Neur/blob/main/dataset')"""
X=mat_contents['data']
n_samples=X.shape[0]

types=mat_contents['types'][0]
weights=mat_contents['weights'][0]

Y=np.zeros((n_samples,10))
typesdict={'P10':10, 'P9':9, 'P8':8,'P7':7,'P6':6,'P5':5,'P4':4,'P3':3,'P2':2,'P1':1}
for i in range(n_samples) :
    orth=typesdict[types[i][0]]-1
    Y[i,orth]=weights[i]

Y_clf=np.ones(Y.shape[0])

for i in range(Y.shape[0]):
    for j in range(Y.shape[1]):
        if(Y[i,j]!=0):
            Y_clf[i]=j

scaler=StandardScaler()
X_clf=scaler.fit_transform(np.abs(X))

C_range = np.logspace(-2, 10, 13)
print(C_range)
gamma_range = np.logspace(-9, 3, 13)
print(gamma_range)
param_grid = dict(gamma=gamma_range, C=C_range)
print('Start gridsearch')
cv = LeaveOneOut()
grid = GridSearchCV(svm.SVC(), param_grid=param_grid, cv=cv)
grid.fit(X_clf, Y_clf)
print(grid)
print("The best parameters are %s with a score of %0.3f (scaled : %0.3f)"
      % (grid.best_params_, grid.best_score_,scaled_score(grid.best_score_,n_class)))

scores = [x[1] for x in grid.cv_results_]
f = 'p' * 67 
for i in f:
  scores.append(i)
print(len(scores)) 

print(scores)
scores = np.array(scores).reshape(len(C_range), len(gamma_range))
img = np.asarray(Image.open('p1.png'))
"""CA=(np.max(scores)-np.min(scores))/2"""
CA = 10 
c = mcolors.ColorConverter().to_rgb
rvb = make_colormap(
            [c('darkblue'), c('blue'), 0.25*CA, c('blue'),c('m'), 0.50*CA, c('m'),c('crimson'),0.66*CA ,c('crimson'),c('darkorange'),0.82*CA ,c('darkorange'),c('gold'),0.91*CA,c('gold'),c('white'),0.96*CA,c('white')])

class MidpointNormalize(Normalize):

    def __init__(self, vmin=None, vmax=None, midpoint=None, clip=False):
        self.midpoint = midpoint
        Normalize.__init__(self, vmin, vmax, clip)

    def __call__(self, value, clip=None):
        x, y = [self.vmin, self.midpoint, self.vmax], [0, 0.5, 1]
        return np.ma.masked_array(np.interp(value, x, y))

plt.figure(figsize=(8, 6))
plt.subplots_adjust(left=.2, right=0.95, bottom=0.15, top=0.95)
plt.imshow(img, interpolation='nearest', cmap=rvb,
           norm=MidpointNormalize(vmin=0.2, midpoint=0.92))
plt.xlabel('gamma')
plt.ylabel('C')
plt.colorbar()
plt.xticks(np.arange(len(gamma_range)), gamma_range, rotation=45)
plt.yticks(np.arange(len(C_range)), C_range)
plt.title('Validation accuracy')
plt.show()
```
